# Exploratory Data Analysis Summary

```python
sns.countplot(x='gender', hue='Churn', data=df)
plt.title("Gender vs Churn")
plt.show()
```

```python
#Observation : Here Female and Male customers have almost similar churn rates.
#Business Insight : Gender is not a major factor influencing customer churn.
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Contract Type vs Churn")
plt.show()
```

```python
#Observation : Month-to-month customers have significantly higher churn.
#Business Insight : Customers with long-term contracts are more loyal.
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='SeniorCitizen', hue='Churn', data=df)
plt.title("SeniorCitizen vs Churn")
plt.show()
```

```python
#Observation :  SeniorCitizen churn rate is relatively higher than that of non-senior customers.
#Business Insight : Senior citizens may require better customer support, simplified plans, or special offers.
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Partner', hue='Churn', data=df)
plt.title("Partner vs Churn")
plt.show()
```

```python
# Observation :  Customers without partners are more likely to churn than those with partners.
# Business Insight : Customers with partners may have more stable subscriptions.
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Dependents', hue='Churn', data=df)
plt.title("Dependents vs Churn")
plt.show()
```

```python
# Observation :  Customers without dependents show a higher churn rate,Customers with dependents are generally more loyal.
# Business Insight : Family-oriented customers tend to remain with the company longer,Introducing family plans may encourage customer retention.
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title("Contract Type vs Churn")
plt.show()
```

```python
# Observation :  Month-to-month contract customers have the highest churn,One-year and two-year contract customers show significantly lower churn.
# Business Insight : Contract duration is one of the strongest indicators of customer retention,Encouraging customers to switch to long-term contracts can substantially reduce churn.
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='InternetService', hue='Churn', data=df)
plt.title("Internet Service vs Churn")
plt.show()
```

```python
# Observation :  Customers using Fiber Optic internet experience the highest churn,DSL users churn less frequently,Customers without internet service have the lowest churn.
# Business Insight : Fiber Optic users may be dissatisfied with pricing, service quality, or support,Improving Fiber Optic service quality and offering loyalty benefits could reduce churn.
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='PaymentMethod', hue='Churn', data=df)
plt.title("Payment Method vs Churn")
plt.xticks(rotation=20)
plt.show()
```

```python
# Observation :  Customers using Electronic Check exhibit the highest churn,Customers paying through automatic payment methods churn less.
# Business Insight : Automatic payment methods appear to improve customer retention,Encouraging customers to adopt automatic payments could reduce churn.
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['tenure'], bins=30, kde=True)
plt.title("Customer Tenure Distribution")
plt.xlabel("Tenure (Months)")
plt.ylabel("Frequency")
plt.show()
```
